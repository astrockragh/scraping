```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs4
import urllib.request
import sys
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
page = urllib.request.urlopen(url)
```

```python
soup = bs4(page, "lxml")
```

```python
print(soup.prettify())
```

```python
right_table=soup.find('table', class_='wikitable sortable')
# right_table
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==6:
        A.append(cells[0].find(text=True))
        mlnk=cells[1].findAll('a')
        contents=str(mlnk[0].contents[0])
        print(type(contents))
        B.append(contents)
        C.append(str(cells[2].find(text=True)))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
```

```python
countrypop=pd.DataFrame(A,columns=['rank'])
countrypop['country']=B
countrypop['population']=C
```

```python
countrypop['population']
```

```python
df=countrypop.astype({'rank': str, 'country': str, 'population': str})
df.population.apply(lambda x: float(x.replace(',', '')))

df.to_pickle("pop.dat")
```
